## Whale Returns

Read the Whale Portfolio daily returns and clean the data

```python
# Reading whale returns
whale_returns_csv = Path("Resources/whale_returns.csv")
# YOUR CODE HERE
whale_returns=pd.read_csv(whale_returns_csv, index_col="Date", infer_datetime_format=True, parse_dates=True)
whale_returns.head(5)
```

## Algorithmic Daily Returns

Read the algorithmic daily returns and clean the data

```python
# Reading algorithmic returns
algo_returns_csv = Path("Resources/algo_returns.csv")
# YOUR CODE HERE
algo_returns=pd.read_csv(algo_returns_csv,index_col='Date', infer_datetime_format=True, parse_dates=True)
algo_returns.head(5)
```

```python
# Count nulls
# YOUR CODE HERE
algo_returns.isnull().sum()
```

```python
# Drop nulls
# YOUR CODE HERE
algo_returns.dropna(inplace=True)
algo_returns.isnull().sum()
```

## S&P 500 Returns

Read the S&P500 Historic Closing Prices and create a new daily returns DataFrame from the data. 

```python
# Reading S&P 500 Closing Prices
sp500_history_csv = Path("Resources/sp500_history.csv")
# YOUR CODE HERE
sp500_history=pd.read_csv(sp500_history_csv, index_col='Date', infer_datetime_format=True, parse_dates=True)
sp500_history.sort_index(inplace=True)
sp500_history.head(5)
```

```python
# Check Data Types
# YOUR CODE HERE
sp500_history.dtypes
```

```python
# Fix Data Types
# YOUR CODE HERE
sp500_history['Close']=sp500_history['Close'].str.replace('$','')
sp500_history['Close']=sp500_history['Close'].astype('float')
sp500_history.dtypes
```

```python
# Calculate Daily Returns
# YOUR CODE HERE
sp500_returns=sp500_history.pct_change()
sp500_returns.head(5)
```

```python
# Drop nulls
# YOUR CODE HERE
sp500_returns.dropna(inplace=True)
sp500_returns.head(5)
```

```python
# Rename Column
# YOUR CODE HERE
sp500_returns.rename(columns={'Close':'S&P 500'}, inplace=True)
sp500_returns.head(5)
```

## Combine Whale, Algorithmic, and S&P 500 Returns

```python
# I check start end dates for each portfolio
#  when i'll do the inner join i'll drop all the data that is not in all the data frames. 
print(f"Whale Returns: \n First row: \n{whale_returns.iloc[0]}; \n \n Last row: \n{whale_returns.iloc[-1]} \n\n\n")
print(f"Algo Returns: \n First row: \n{algo_returns.iloc[0]}; \n \n Last row: \n{algo_returns.iloc[-1]} \n\n\n")
print(f"S&P 500 Returns: \n First row: \n{sp500_returns.iloc[0]}; \n \n Last row: \n{sp500_returns.iloc[-1]} \n\n\n")
```

```python
# Concatenate all DataFrames into a single DataFrame
# YOUR CODE HERE
combined_returns=pd.concat([whale_returns, algo_returns, sp500_returns], axis='columns', join='inner')
combined_returns.head(5)
```

```python
# Confirm start / end date are as expected:
print(f"Combined Returns: \n First row: \n{combined_returns.iloc[0]}; \n \n Last row: \n{combined_returns.iloc[-1]} \n\n\n")
```

## Rolling Statistics

Risk changes over time. Analyze the rolling statistics for Risk and Beta. 

1. Plot the rolling standard deviation of the various portfolios along with the rolling standard deviation of the S&P 500 (consider a 21 day window). Does the risk increase for each of the portfolios at the same time risk increases in the S&P?
2. Construct a correlation table for the algorithmic, whale, and S&P 500 returns. Which returns most closely mimic the S&P?
3. Choose one portfolio and plot a rolling beta between that portfolio's returns and S&P 500 returns. Does the portfolio seem sensitive to movements in the S&P 500?
4. An alternative way to calculate a rolling window is to take the exponentially weighted moving average. This is like a moving window average, but it assigns greater importance to more recent observations. Try calculating the ewm with a 21 day half-life.

```python
# Calculate and plot the rolling standard deviation for
# the S&P 500 and whale portfolios using a 21 trading day window
# YOUR CODE HERE
xmin=combined_returns.index.min()
xmax=combined_returns.index.max()
combined_returns.rolling(window=21).std().plot(figsize=(20,10), title='21 Day Rolling Standard Deviation',xlim=(xmin,xmax))
print("It's hard to tell from the chart, but all the portfolios seem to move in tandem with the S&P 500, both when the standard deviation increases as well as when it decreases. Some portfolios, Tiger Global Management and Berkshire Hathawat, have more pronounced swings than the S&P 500 and other portfolios")
```

```python
# Calculate Beta for a single portfolio compared to the total market (S&P 500)
# (Your graph may differ, dependent upon which portfolio you are comparing)
# YOUR CODE HERE

# I determine the rolling timeframe (60 seems to be the same used in the original chart), the portfolio I want the beta for
rolling_window=60
portfolio_name='BERKSHIRE HATHAWAY INC'
covariance=combined_returns[portfolio_name].rolling(window=rolling_window).cov(combined_returns['S&P 500'])

# I calculate the variance of the S&P 500 that will be the denominator of the beta calculation
variance_sp500=combined_returns['S&P 500'].rolling(window=rolling_window).var()

# I plot the chart of the rolling Beta for the portfolio I chose
xmin=(covariance/variance_sp500).index.min()
xmax=(covariance/variance_sp500).index.max()
(covariance/variance_sp500).plot(figsize=(20,10),xlim=(xmin,xmax), title=f"{portfolio_name} Beta")
```

```python
# Calculate a rolling window using the exponentially weighted moving average. 
# YOUR CODE HERE
# the function for the exponential weighted average is "ewm()".
combined_returns.ewm(halflife=21).std().plot(figsize=(20,10), title='Exponetially Weighted Average (21 days)')
```

## Sharpe Ratios
In reality, investment managers and thier institutional investors look at the ratio of return-to-risk, and not just returns alone. (After all, if you could invest in one of two portfolios, each offered the same 10% return, yet one offered lower risk, you'd take that one, right?)

1. Using the daily returns, calculate and visualize the Sharpe ratios using a bar plot.
2. Determine whether the algorithmic strategies outperform both the market (S&P 500) and the whales portfolios.

```python
# Visualize the sharpe ratios as a bar plot
# YOUR CODE HERE
risk_free=0
((combined_returns.mean()*252-risk_free)/(combined_returns.std()*np.sqrt(252))).sort_values().plot(kind='bar', title='Sharpe Ratios')
```

```python
print(f"Algo 1 strategy outperforms, when measured by its sharpe ratios, the S&P 500 as well as any other portfolio under analysis. On the other hand, Algo 2, underperformns relative to Berjshire Hathaway, S&P 500 and Algo 1")
```

```python
# Read the first stock
# YOUR CODE HERE
csv_file=Path("Resources/goog_historical.csv")
google_df=pd.read_csv(csv_file, index_col='Trade DATE', infer_datetime_format=True, parse_dates=True)
google_df.sort_index(ascending=True, inplace=True)
google_df.head(5)
```

```python
# Read the second stock
# YOUR CODE HERE
csv_file=Path("Resources/aapl_historical.csv")
apple_df=pd.read_csv(csv_file, index_col='Trade DATE', infer_datetime_format=True, parse_dates=True)
apple_df.sort_index(ascending=True, inplace=True)
apple_df.head(5)
```

```python
# Read the third stock
# YOUR CODE HERE
csv_file=Path("Resources/cost_historical.csv")
costco_df=pd.read_csv(csv_file, index_col='Trade DATE', infer_datetime_format=True, parse_dates=True)
costco_df.sort_index(ascending=True, inplace=True)
costco_df.head(5)
```

```python
# Concatenate all stocks into a single DataFrame
# YOUR CODE HERE
# Concatenate these 3 data frames by rows, i.e. one on top of the other. 
combined_new=pd.concat([google_df,apple_df,costco_df], join='inner',axis='rows')
combined_new.head()
```

```python
# Reset the index
# YOUR CODE HERE
combined_new=combined_new.reset_index()
combined_new.head(5)
```

```python
# Pivot so that each column of prices represents a unique symbol
# YOUR CODE HERE
combined_new=combined_new.pivot_table(values='NOCP', index='Trade DATE', columns='Symbol')
combined_new.head(5)
```

```python
# Drop Nulls
# YOUR CODE HERE
# I calculate the daily returns and drop Nulls
combined_new_returns=combined_new.pct_change()
combined_new_returns.dropna(inplace=True)
combined_new_returns.head(5)
```

## Calculate the weighted returns for the portfolio assuming an equal number of shares for each stock

```python
# Calculate weighted portfolio returns
weights = [1/3, 1/3, 1/3]
# YOUR CODE HERE
new_portfolio=combined_new_returns.dot(weights)
new_portfolio.rename('Custom', inplace=True)
new_portfolio.head(5)
```

## Join your portfolio returns to the DataFrame that contains all of the portfolio returns

```python
# Add your "Custom" portfolio to the larger dataframe of fund returns
# YOUR CODE HERE
combined_returns=pd.concat([combined_returns,new_portfolio],join='inner',axis='columns')
combined_returns.head(5)
```

```python
# Only compare dates where return data exists for all the stocks (drop NaNs)
# YOUR CODE HERE
# check for NaNs: there seems to be none
combined_returns.isnull().mean()
```
